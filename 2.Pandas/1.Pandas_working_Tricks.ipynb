<h1 style="color: Gold;font-size:200%; font-weight:bold">
1)Selecting Columns by Data Types</h1>

<p style="color: magenta;font-size:120%; font-weight:bold; margin-left: 50px;" >
USE CASE:: If youâ€™re performing statistical analysis or ML then you need to isolate only the numerical column
in that time it become too handly </p>

```python
#Example: Lets take data from searborn 
import numpy as np 
import pandas as pd 
import seaborn as sns 

# Load the Titanic dataset
df = sns.load_dataset('titanic')
# df.info() #checking the info/datatypes raw data 

#Now Select only numerical columns (int and float data types)
df_numerical = df.select_dtypes(include=['int', 'float'])
df_numerical.info()
```

<h1 style="color: Gold;font-size:200%; font-weight:bold">
2)Conditional Filtering with query</h1>

<p style="color: magenta;font-size:120%; font-weight:bold; margin-left: 50px;" >
USE CASE: Become useful when you need filter data based on multiple conditions, such as selecting rows where the age is greater than 25 and embark_town is "Cherbourg" in titanic data</p>

```python
import pandas as pd 
import seaborn as sns 
df=sns.load_dataset('titanic')

df_filtered=df.query('age>25 & embark_town=="Cherbourg"')
df_filtered.sample(5)
```

<h1 style="color: Gold;font-size:200%; font-weight:bold">
3) Memory Optimization with astype</h1>

<ul style="color: magenta;font-size:120%; font-weight:bold; margin-left: 50px;" >
<li>It is used to optimize memory usage by converting columns to more efficient types, such as converting integers to int32 or categoricals to the category type.</li>
<br>
<li>USE CASE: When you are working in large dataset you may need to reduce its memory, this can be done by converting categorical--> category type OR large integers --> smaller integer types
</li> </br></ul>

```python
import pandas as pd 
import seaborn as sns 
df=sns.load_dataset('titanic')

#Integer to Int16 
df_small_numerical=df.select_dtypes(include=['int']).astype('int16')
df_small_numerical.info()

#categogorical to category 
df_small_categorical=df.select_dtypes(include=['object']).astype('category')
df_small_categorical.info()
```
