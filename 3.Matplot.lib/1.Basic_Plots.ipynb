Basic Plots in pyplot interface</h1>

<h1 style=" color:gold; font-size:200%; font-weight:bold"> 1)Line plot </h1>
<ul>
    <li style="color:crimson;font-weight:bold">General Syntax</li>
         <ul style="color: violet; font-weight:bold">plt.plot(x, y, fmt, **kwargs) </ul>
</ul>

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate different x values for two datasets
x1 = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x1)   # Corresponding y values
x2 = np.linspace(0, 5, 50)    # 50 points from 0 to 5
y2 = np.cos(x2)     # Corresponding y values

# Plot both lines on the same graph with different x values
plt.plot(x1, y1, 'r-', x2, y2, 'b--')

plt.title('Plot with Different x Values')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()  # Display legend
plt.show()
```

### if data is : Dataframe 

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a simple DataFrame
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Revenue': [10, 15, 12, 20, 18]}

df = pd.DataFrame(data)
df
# Use the 'data' parameter to plot
plt.plot('Year', 'Revenue', data=df, marker='o', color='green')
plt.title('Yearly Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue (in millions)')
plt.show()
```

<h1 style=" color:gold; font-size:200%; font-weight:bold"> 2)Scatter Plot </h1>
<ul>
    <li style="color:crimson;font-weight:bold">General Syntax</li>
         <ul style="color: violet; font-weight:bold">plt.scatter(x, y, **kwargs) </ul>
</ul>

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x1 = np.random.rand(50)  # 50 random points for x
y1 = np.random.rand(50)  # 50 random points for y
x2 = np.random.rand(40)  # 40 random points for x
y2 = np.random.rand(40)  # 40 random points for y

# Create scatter plot
plt.scatter(x1, y1, color='b', label='1st pair')  
plt.scatter(x2, y2,  color='r', label='2nd pair')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend()
plt.show()
```

<h1 style=" color:gold; font-size:200%; font-weight:bold"> 3)Bar Plot </h1>
<ul>
    <li style="color:crimson;font-weight:bold">General Syntax</li>
         <ul style="color: violet; font-weight:bold">
            <li>Vertical bar plot:=======>            plt.bar(x, y, **kwargs)</li>
            <li>Horizontal bar plot:======>          plt.barh(x, y, **kwargs)</li>  
         </ul>
</ul>

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Create bar chart
plt.bar( categories,values,  color='blue', width=0.5)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical bar chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Create bar chart
plt.barh( categories,values, color='pink', height=0.5)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Horizontal Bar Chart')
plt.show()
```
